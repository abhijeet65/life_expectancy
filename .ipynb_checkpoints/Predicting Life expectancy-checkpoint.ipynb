### A simple Linear Regression Model that predicts the life expectancy from the BMI(Body Mass Index) of a person.

```python
# Import the libraries
# because the cotributors are awesome
import numpy as np
import pandas as pd
from sklearn.linear_model import LinearRegression
import matplotlib.pyplot as plt
%matplotlib inline
```

```python
# Get the dataset
bmi_life_data = pd.read_csv("/Users/apple/Desktop/bmi_and_life_expectancy.csv")
bmi_life_data.head()
```

```python
## Lets get the variables

# The independent variable x, because it has got wings to fly and no boundaries
x = bmi_life_data[['BMI']]

# The dependent variable y, because it wants to stalk x. Creepy AF! 
y = bmi_life_data[['Life expectancy']]
```

```python
plt.scatter(x.values,y.values)
plt.xlabel("BMI")
plt.ylabel("Life Expectancy")
```

```python
# Plot it 
model = LinearRegression()
model.fit(x,y)
```

```python
# Let's see the line that fits into this data. 
# alpha and c are just to make it colorful, no one wants to be dull.

plt.scatter(x.values,y.values, alpha=0.5, c=x.values)
plt.xlabel("BMI")
plt.ylabel("Life Expectancy")
plt.plot(x, model.predict(x), color='red', linewidth=1)
plt.show()
```

```python
# Now, the fun part.
# I've got few friends, lets get their BMIs

BMI = { "Ashwani" : 23.3,
          "Himanshu" : 22,
          "Abhinav" : 21.7}

# Oh! Ankit always arrives late to the party.
BMI["Ankit"] = 26

for name, bmi in BMI.items():
    print("The BMI of ", name , "is", bmi, " and the life expectancy is ", model.predict(bmi))
```
